```python
#Visualization with pie chart using Matplotlib 

from collections import Counter

matplotlib.style.use('ggplot')

room = df1.room_type
r = Counter(room)

room_count = pd.DataFrame.from_dict(r, orient='index').sort_values(by=0)
room_count.columns = ['room_type']
room_count.plot.pie(y = 'room_type', 
                 colormap = 'plasma', 
                 figsize=(10,10), 
                 fontsize = 20, autopct = '%.2f',
                 legend = False,
                 title = 'Distribution of the type of room')
```
